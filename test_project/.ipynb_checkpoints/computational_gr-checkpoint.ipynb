```python
from IPython.display import display
from sympy import *
from gravipy import *

init_printing()

t,x, y, z = symbols("t,x,y,z ")
C = Coordinates('\chi',[t,x,y,z])

A = Function('A')(z)
### PRINT METRIC ###
Metric = diag(-A**2, A**2, A**2, 1)
g = MetricTensor('g',C,Metric)
print("Metric is : ")
display(g(All,All))
print("\n")


### COMPUTE CHRISTOFFEL SYMBOL ###
Ga = Christoffel('Ga',g)
#print Ga.components      ----- This command is not working for me
print("Christoffel symbol is : ")
display(Ga(-All,All,All))
print("\n")

### COMPUTE RIEMANN TENSOR ###
Rm = Riemann('Rm',g)
print("Riemann tensor is : ")
display(Rm(-All,All,All,All))
print("\n")


### COMPUTE RICCI TENSOR AND RICCI SCALAR ###
Ri = Ricci('Ri',g)
print("Ricci tensor is : ")
display(Ri(All, All))
print("\n")
display("Ricci scalar is : " , Ri.scalar())
print("\n")

```
